```python
import pandas as pd
import numpy as np
import tqdm
from tqdm import tqdm
from tqdm import trange
```

```python
data = pd.read_csv("jnd_sent.csv")
data
```

```python
def keywords(data):
    from jieba import analyse
    textrank = analyse.textrank
    keywords = textrank(data, topK=8)
    return ' '.join(keywords)
```

```python
data['关键词'] = [keywords(str(i)) for i in tqdm(list(data['评论']))]
```

```python
#data.drop('Unnamed: 0',axis=1, inplace=True)
#data.to_csv('textrank结果.csv', index=False)
milk = data[data['res']==-1]
```

```python
milk
```

```python
keylst = []
for i in trange(len(milk)):
    key = milk.iloc[i,8]
    keys = key.split(' ')
    keylst = keylst + keys
# keylst = keylst.remove('')
keylst
```

```python
keyset=set(keylst)
keydict={}
for item in keyset:
    keydict.update({item:keylst.count(item)})
print(keydict)
```

```python
d_order=sorted(keydict.items(),key=lambda x:x[1],reverse=True)
d_order
```
