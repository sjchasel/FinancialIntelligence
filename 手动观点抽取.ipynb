```python
import pandas as pd
import numpy as np
import tqdm
from tqdm import tqdm
from tqdm import trange
import re
```

```python
data = pd.read_csv("alldata_sent.csv")
jiangnandao = data[data['店铺名称']=='江南道自助烤肉海鲜涮锅（柳浪湾店）']
jiangnandao.index = range(len(jiangnandao))
jiangnandao
```

```python
#味道
taste = ['好吃','味道','难吃','新鲜','不好吃','发苦','口味','酸']
# 环境
environ = ['环境','挤','卫生','吵','脏','乱','差']
# 服务
serve = ['服务','态度']
# 性价比
price = ['性价比','便宜','贵','价格','物美价廉']
# 等位
deng = ['等位','排队','慢','等','爆满','预约']
```

```python
jiangnandao['观点'] = ''
for i in trange(len(jiangnandao)):
    aspect = []
    comment = jiangnandao['评论'][i]
    for t in taste:
        if t in comment:
            aspect.append('味道')
            break
    for e in environ:
        if e in comment:
            aspect.append("环境")
            break
    for s in serve:
        if s in comment:
            aspect.append('服务')
            break
    for s in price:
        if s in comment:
            aspect.append('性价比')
            break
    for s in deng:
        if s in comment:
            aspect.append('等位')
            break
    jiangnandao['观点'][i] = aspect
    
jiangnandao[jiangnandao['观点']!='']
```
