```python
import pandas as pd
import numpy as np
import tqdm
from tqdm import tqdm
from tqdm import trange
from sentiment import *
```

```python
alldata = pd.read_csv('全部评论数据.csv')
name_cate = pd.read_excel('name_cate.xlsx')
name_cate
```

```python
cates = ['火锅',
 '东南亚菜',
 '烧烤烤肉',
 '小吃快餐',
 '川湘菜',
 '西餐',
 '聚餐宴请',
 '自助餐',
 '甜点饮品',
 '京菜鲁菜',
 '日韩料理',
 '蛋糕甜品',
 '汤锅',
 '香锅烤鱼']
```

```python
dictt = {i:[] for i in cates}
for i in range(len(name_cate)):
    cate = name_cate.iloc[i,1]
    for j in cates:
        if j in cate:
            dictt[j].append(name_cate.iloc[i,0])
dictt
```

```python
len(alldata['店铺名称'].unique())
```

```python
for cate in cates:
    names = dictt[cate]
    datas = []
    for name in names:
        datas.append(alldata[alldata['店铺名称']==name].drop_duplicates(subset=['店铺名称','用户名','评论'],keep='first'))
    data = pd.concat(datas).reset_index(drop=True)
    data.to_csv(cate+'.csv',encoding="utf-8-sig")
```

```python
alldatanew = pd.read_csv('全部评论数据.csv')
sent = Sentiment(alldatanew)
scores, result = sent.run()
alldatanew['scores'] = scores
alldatanew['res'] = result
alldatanew.to_csv('alldata_sent.csv',encoding="utf_8_sig")
```

```python
keydict = {}
for name in list(name_cate['店铺名']):
    data = alldatanew[alldatanew['店铺名称']==name]
    pos = list(data['res']).count(1)
    neg = list(data['res']).count(-1)
    score = pos / (pos + neg)
    cate = str(name_cate[name_cate['店铺名']==name]['类别'])
    keydict[name] = score
    print('店铺名：{},所属类别：{},正向评论数量：{}，负向评论数量：{}，情感得分（正向评论数/评论总数）：{}'.format(str(name),str(cate),str(pos),str(neg),str(score)))
d_order=sorted(keydict.items(),key=lambda x:x[1],reverse=True)
d_order
```
